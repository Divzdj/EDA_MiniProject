```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data=pd.read_csv('IPL_Dataset.csv')
```

```python
data
```

## Q1. What is the maximum number of matches played by an individual player in a season?

```python
data.head()
```

```python
max_matches = data.groupby('Player')['Mat'].max().max()

print("Maximum number of matches played by an individual player in a season:", max_matches)
```

##  Q2. What is the highest average of a player in a season?

```python
highest_average=data['Avg'].max()
```

```python
players_with_highest_average=data[data['Avg']==highest_average]['Player']
print("Highest average of a player:",highest_average)
print("Player with the highest average:",players_with_highest_average)
```

## Q3. What is the average number of matches played by the players in a season?

```python
average_matches = data['Mat'].mean()
print("Average number of matches played by players in a season:", average_matches)
```

## Q4. Please check the correlation between all the features.

```python

correlation_matrix = data.corr()
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', square=True)
plt.title('Correlation Matrix Heatmap')
plt.show()
```

## Q5. Check the list of players who has an average greater than 50 as well strike rate above 120.

```python
filtered_data = data[(data['Avg'] > 50) & (data['SR'] > 120)]
players_list = filtered_data['Player'].tolist()
print("List of players with an average greater than 50 and a strike rate above 120:")
for player in players_list:
    print(player)
```

## Q6. Please check the list of players who has an average greater than 40 and balls faced above 100.

```python
fil_data=data[(data['Avg']>40)&(data['BF']>100)]
players_list=fil_data['Player'].tolist()
print("List of players who has an average greater than 40 and balls faced above 100:")
for player in players_list:
    print(player)
```

## Q7. Please check the list of players who scored more than 300 runs with an average greater than 45 & strike rate above 130.

```python
f_data=data[(((data['Runs']>300)&(data['Avg']>45))&(data['SR']>130))]
players_list=f_data['Player'].tolist()
print("List of players who scored more than 300 runs with an average greater than 45 & strike rate above 130:")
for player in players_list:
    print(player)
```

## Q8. Players who scored alteast one century in this season.

```python
fil_data=data[data['100']>=1]
players_list=fil_data['Player'].tolist()
print("Players who scored alteast one century in this season:")
for player in players_list:
    print(player)
```

## Q9. Players who scored atleast 4 half centuries in this season.

```python
filt_data=data[data['50']>=4]
players_list=filt_data['Player'].tolist()
print("Players who scored atleast 4 half centuries in this season:")
for player in players_list:
    print(player)
```

##  Q10. Check the list of players who hit more than 45 boundaries and more than 10 sixes in this season.

```python
fil_data=data[(data['4s']>45)&(data['6s']>10)]
players_list=fil_data['Player'].tolist()
print("List of players who hit more than 45 boundaries and more than 10 sixes in this season:")
for player in players_list:
    print(player)
```

## Q11. Plot a histogram of number of matches played in a season by players.

```python
player_matches=data[['Player','Mat']]
plt.figure(figsize=(10,6))
plt.hist(player_matches['Mat'],bins=20,edgecolor='black')
plt.xlabel("Number of Matches")
plt.ylabel('Frequency')
plt.title('Histogram of Number of Matches Played in a Season by Players')
plt.show()
```

## Q12. Plot the histogram of balls faced by players.

```python
balls_faced=data['BF']
plt.figure(figsize=(10, 6))
plt.hist(balls_faced,bins=20,edgecolor='black')
ax = plt.gca()
ax.set_xlabel('No. of Balls Faced')
ax.set_ylabel('Frequency')
plt.title('Histogram of Number of Balls Faced by Players')
plt.show()
```

## Q13. Top 20 players with most runs in a season.

```python
sorted_data=data.sort_values(by='Runs',ascending=False)
top_20_players=sorted_data.head(20)
print("Top 20 players with the most runs in a season:")
print(top_20_players[['Player', 'Runs']])
```

## Q14. Bar chart for top 20 players with best strike rate in a season.

```python

sorted_data=data.sort_values(by='SR',ascending=False)
top_20_players=sorted_data.head(20)
plt.figure(figsize=(12, 6))
plt.bar(top_20_players['Player'], top_20_players['SR'])
plt.title('Top 20 Players with Best Strike Rate in a Season')
plt.xticks(rotation=90)
plt.show()
```

## Q15. Top 30 players with highest number of sixes.

```python
sorted_values=data.sort_values(by='6s',ascending=False)
top_30_players=sorted_values.head(30)
plt.figure(figsize=(10,6))
plt.bar(top_30_players['Player'],top_30_players['6s'])
plt.title('Top 30 players with highest number of sixes')
plt.xticks(rotation=90)
plt.show()
```

## Q16. Scatter plot of runs scored by a player v/s balls faced in a season.

```python
runs_balls_data = data[['Runs', 'BF']]
plt.figure(figsize=(10, 6))
plt.scatter(runs_balls_data['BF'], runs_balls_data['Runs'])
ax = plt.gca()
ax.set_xlabel('Balls Faced')
ax.set_ylabel('Runs')
plt.title('Runs Scored vs. Balls Faced')
plt.show()
```

## Q17. KDE plot for averages on a season.

```python
averages_data = data['Avg']
plt.figure(figsize=(10, 6))
ax=sns.kdeplot(averages_data, shade=True)
ax.set(xlabel='Average', ylabel='Density')

plt.title('KDE Plot of Averages in a Season')
plt.show()
```
